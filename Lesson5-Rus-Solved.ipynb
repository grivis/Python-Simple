## Изучим эту проблему! ##

Напишем программу, которая берет текст и перепутывает буквы в словах. И посмотрим, насколько сильно можно перепутать буквы, чтобы текст все еще оставался читаемым...

***Начинаем писать программу*...**

Прежде всего, создадим переменную, которая содержит фрагмент текста:

```python
text = '''Как уже всем известно, Незнайка после путешествия значительно поумнел, стал учиться читать и писать, прочитал всю грамматику и почти всю арифметику, стал делать задачки и уже даже хотел начать изучать физику, которую в шутку называл физикой-мизикой, но как раз тут ему почему-то расхотелось учиться. Это часто случается в стране коротышек. Иной коротышка наобещает с три короба, наговорит, что сделает и это и то, даже горы свернет и вверх ногами перевернет, на самом же деле поработает несколько дней в полную силу, а потом снова понемножку начинает отлынивать.'''
```

Пока не видим никакого результата. Вот если напечатать текст, то увидим результат...

```python
print(text)
```

Но нам нужен не весь текст целиком, а каждое слово в отдельности... Как разбить текст на слова?

```python
textwords = text.split()
print(textwords)
```

Мы получили список слов, из которых состоит текст. Видим, что вместе со словами здесь есть и знаки препинания. Они будут мешать нашей программе. Как удалить знаки препинания? 

```python
text = text.replace('.', '')
text = text.replace(',', '')
textwords = text.split()
print(textwords)
```

А теперь нам нужно написать такой код на Питоне, который возьмет каждое слово, выделит первую и последнюю буквы (оставив их на своем месте), но переставит местами буквы посередине. Например, вывернет среднюю часть слова "наизнанку". Вот так: Ф ОРМ А --> Ф МРО А

Для этого удобно использовать срезы. Например:

```python
a = [1, 2, 3, 4, 5]
print( a[::-1])
print(a[1:4:2])
print(a[::-1])
```

```python
for word in textwords:
    fl = word[0]
    ll = word[-1:]
    middle = word[-2:0:-1]
    neword = fl + middle + ll
    print(neword, end=' ')
```

Получилось... Но текст читается очень плохо... Значит, выворачивание середины слова "наизнанку" - это слишком сильное изменение. Нужна другая программа. Например, поменять местами всего две буквы в середине слова... При этом надо учесть, что слова короче трех символом изменить таким образом невозможно. А слова длиной четыре символа меняются только одним способом.

```python
for word in textwords:
    if len(word)>3:
        fl = word[0]
        sl = word[1]
        tl = word[2]
        rest = word[3:]
        neword = fl + tl + sl + rest
    else:
        neword = word
    print(neword, end=' ')
```

Теперь получается слишком просто. Изменим нашу программу. Теперь надо менять местами две случайные буквы в середине слова. Не забудем подключить нужный модуль для получения случайных чисел.

```python
import random
for word in textwords:
    if len(word)>4:
        fl = word[0]
        ll = word[-1:]
        middle = word[1:-1]
        x = random.randint(0, len(middle)-2)
        midlist = list(middle)
        midlist[x], midlist[x+1] = midlist[x+1], midlist[x]
        middle = ''
        for i in midlist:
            middle += i
        neword = fl + middle + ll
    else:
        neword = word
    print(neword, end=' ')
```

А теперь перепишем код. Создадим функцию, которая получает очередное слово, и возвращает это же слово после преобразования...

```python
def shuffle(word):
    if len(word)>4:
        fl = word[0]
        ll = word[-1:]
        middle = word[1:-1]
        x = random.randint(0, len(middle)-2)
        midlist = list(middle)
        midlist[x], midlist[x+1] = midlist[x+1], midlist[x]
        middle = ''
        for i in midlist:
            middle += i
        neword = fl + middle + ll
    else:
        neword = word
    return neword
    
```

А теперь напишем основную программу, которая в цикле вызывает написанную нами только что функцию...

```python
import random
for word in textwords:
    print(shuffle(word), end=' ')
```

А теперь надо написать код, который соберет отдельные слова в слитный текст опять. И распечатает его...

```python
newtext = ''
import random
for word in textwords:
    newtext += shuffle(word) + ' '
print(newtext)
```

## Получилось! ##

Мы видим, что текст с переставленными буквами читать можно... Но чем больше букв переставлено, и чем дальше переставленные буквы отстоят друг от друга, тем сложнее читать получившийся текст...

## А теперь - совсем другая задача! ##

Как сделать текст **совсем нечитаемым**? То есть, **закодировать** его... Разумеется мы хотим, чтобы текст все же смогли прочитать наши друзья, которые знают, как **расшифровать его**...

Самый простой путь - заменить одни буквы на другие... А потом, при расшифровке, вернуть все буквы на место.

Создадим словарь для кодирования:

```python
encodic = {'а':'э', 'б':'ю', 'в':'я'}
```

Поскольку наш словарь - неполный, т.е. в нем не все буквы, надо предусмотреть, чтобы буквы, для которых нет замены, оставались на своеем месте...

Итак, напишем всю программу целиком: она берет текст, обходит букву за буквой и заменяет буквы в соответствии со словарем

```python
textlts = list(text)
encodtext = ''
for letter in textlts:
    letter = encodic.get(letter, letter)
    encodtext += letter
print(encodtext)
 
```

Работает! А вот теперь нужна программа, которая делает все наоборот. То есть, раскодирует текст, возвращая буквы на свои места...

```python
decodic = {'э':'а', 'ю':'б', 'я':'в'}
```

```python
textlts = list(encodtext)
decodtext = ''
for letter in textlts:
    letter = decodic.get(letter, letter)
    decodtext += letter
print(decodtext)
```

Улучшим программу... Будем менять буквы на цифры и различные знаки... Изменим словарь.
